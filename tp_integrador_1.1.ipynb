```python
def get_dolar_blue():
    url = "https://dolarapi.com/v1/dolares/blue"

    try:
        response = requests.get(url, timeout=10)
        response.raise_for_status()

        data_usd = response.json()
        return data_usd["venta"]  # Suponiendo que "venta" es el valor de venta del dolar blue.

    except requests.RequestException as e:
        print(f"Error fetching Dolar Blue value: {e}")
        return None
```

```python
book_name= soup.find_all(class_= 'name product-title woocommerce-loop-product__title')

url_text= [element.find('a')['href'] for element in book_name]

#Ranking= iD
p_tags = soup.find_all('p', style='font-size: 12px; text-align: center; line-height: 1.3; margin-bottom: 10px; color: white; padding: 3px; background: #f26636; border-radius: 4px; margin-top: 5px')
ranking_list = []

for p_tag in p_tags:
    ranking_text = p_tag.text
    # Utiliza expresión regular para extraer el número de ranking
    match = re.search(r'#(\d+)', ranking_text)
        
    if match:
        ranking_number = int(match.group(1))
        ranking_list.append(ranking_number)
    else:
        print("No se encontró el número de ranking en la etiqueta <p>:", p_tag)

# Verificar los datos antes de agregar a data
print("book_name:", [element.text for element in book_name])
print("url_text:", url_text)
print("ranking_list:", ranking_list)

```

```python
driver = webdriver.Chrome()

converted_price_ars = []
converted_price_usd = []



# ... tu código anterior ...

for book_url in url_text:
    # Abrir la URL en el navegador
    driver.get(book_url)

    try:
        # Esperar durante 5 segundos (ajusta según sea necesario)
        time.sleep(5)

        # Encontrar la etiqueta <p> que contiene el precio en ARS
        price_tag_ars = driver.find_element(By.CSS_SELECTOR, 'p.price.product-page-price span.woocommerce-Price-amount bdi')

        # Extraer el texto del precio en ARS
        price_ars = price_tag_ars.text.strip()
        price_ars = price_ars.replace('$', '').replace('\xa0', '').replace('.', '').replace(',', '.')
        # Convertir a float y agregar a la lista
        converted_price_ars.append(float(price_ars))
        print(f"Precio del libro en AR$ {book_url}: {price_ars}")

    except Exception as e:
        converted_price_ars.append(None)
        print(f"No se pudo obtener el precio en ARS para {book_url}. Error: {str(e)}")

    try:
        # Esperar durante 5 segundos (ajusta según sea necesario)
        time.sleep(5)

        # Encontrar la etiqueta <span> que contiene el precio en USD
        price_tag_usd = driver.find_element(By.CSS_SELECTOR, 'span[style="color: #111; font-weight: 400; font-size: 1.1em"] span[style="font-size: 1.3em"]')

        # Extraer el texto del precio en USD
        price_usd = price_tag_usd.text.strip()
        price_usd = price_usd.replace(',', '.')
        # Convertir a float y agregar a la lista
        converted_price_usd.append(float(price_usd))
        print(f"Precio del libro en U$D {book_url}: {price_usd}")

    except Exception as e:
        converted_price_usd.append(None)
        print(f"No se pudo obtener el precio en USD para {book_url}. Error: {str(e)}")

    try:
        # Esperar durante 10 segundos a que la etiqueta <div> que contiene la categoría esté presente
        category_tag = WebDriverWait(driver, 10).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, 'div.product_meta a[rel="tag"]'))
        )

        # Extraer el texto de la categoría
        category_text = category_tag.text.strip()
        print(f'Categoría del libro {book_url}: {category_text}')

    except Exception as e:
        category_text = None
        print(f"No se pudo obtener la categoría para {book_url}. Error: {str(e)}")

# Cerrar el navegador al finalizar
driver.quit()

# Imprimir las listas convertidas
print("Precios ARS convertidos:", converted_price_ars)
print("Precios USD convertidos:", converted_price_usd)

```
